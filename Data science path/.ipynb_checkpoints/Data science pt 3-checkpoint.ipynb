### implementing KNN

```python
#separate data 
X = df.drop('Cancer Present',axis=1)
y = df['Cancer Present']
```

```python
#train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
#scaling in KNN is a MUST
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print(classification_report(y_test,y_pred))
```

## Pipeline

is an object that allow us to do particular task in sequence. instead of using the same code over and over

```python
from sklearn.metrics import accuracy_score
```

```python
#finding the best K from scratch using iteration

test_error_rates = []

for i in range(1,30):
    knn_model = KNeighborsClassifier(n_neighbors=i)
    knn_model.fit(X_train,y_train)
    
    y_pred = knn_model.predict(X_test)
    
    test_error = 1 - accuracy_score(y_test,y_pred)
    test_error_rates.append(test_error)
```

```python
len(test_error_rates)
```

```python
sns.lineplot(x= range(1,30), y=test_error_rates)
plt.xticks(range(1,30))
plt.xlabel(range(1,30))
plt.show()
```

```python
plt.plot(range(1,30),test_error_rates)
```

based on that plot, the elbow point is around 7

```python
from sklearn.metrics import SCORERS
SCORERS.keys()
```

```python
# pipeline method
# we already have created scaler and our model
knn_model = KNeighborsClassifier()
scaler = StandardScaler()
k_values = list(range(1,21))
```

```python
knn_model.get_params().keys()
```

```python
# set our operation
operation = [('scaler',scaler),('knn',knn_model)]
```

```python
from sklearn.pipeline import Pipeline
# build our pipe
pipe = Pipeline(operation)

#using pipe in gridsearch
param_grid = {'knn__n_neighbors':k_values,'knn__algorithm':['brute','ball_tree'],
             'knn__leaf_size':[30,100,250,500,1000,1500]}
search = GridSearchCV(estimator=pipe, param_grid=param_grid,cv=5, scoring='f1')
```

```python
# we need to reset the value to unscaled
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
search.fit(X_train,y_train)
```

```python
search.best_estimator_
```

```python
y_pred = search.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report

search_score = accuracy_score(y_test,y_pred)
print('acc score:',search_score)
print(classification_report(y_test,y_pred))
```

```python
search.get_params()
```

```python
from sklearn.svm import LinearSVR, SVR
```

```python
df = pd.read_csv("./Jose lesson/DATA/cement_slump.csv")
X = df.drop('Compressive Strength (28-day)(Mpa)',axis=1)
y = df['Compressive Strength (28-day)(Mpa)']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=101)
scaler = StandardScaler()
scld_Xtrain = scaler.fit_transform(X_train)
scld_Xtest = scaler.transform(X_test)
```

```python
#without searching best params
base_svr = SVR()
base_svr.fit(scld_Xtrain,y_train)
default_yhat = base_svr.predict(scld_Xtest)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
def_rmse = np.sqrt(mean_squared_error(y_test,default_yhat))
def_mae = mean_absolute_error(y_test,default_yhat)
def_mae
```

```python
#searching best_params
param_grid = {'C':[0.001,0.01,0.1,0.5,1],
             'kernel':['linear','rbf','poly'],
              'gamma':['scale','auto'],
              'degree':[2,3,4],
              'epsilon':[0,0.01,0.1,0.5,1,2]}
```

```python
search = GridSearchCV(base_svr,param_grid)
search.fit(scld_Xtrain,y_train)
```

```python
search.best_params_
```

```python
search_yhat = search.predict(scld_Xtest)
mean_absolute_error(y_test,search_yhat)
```

```python
df = pd.read_csv("./Jose lesson/DATA/penguins_size.csv")
df.head()
```

```python
def missing_value(df):
    percent = df.isnull().sum() * 100 / len(df)
    percent = percent[percent > 0 ].sort_values()
    return percent
```

```python
df.info()
```

```python
missing_value(df)
```

```python
df.loc[(df.culmen_length_mm.isnull()) & (df.culmen_depth_mm.isnull())]
```

```python
df.dropna(subset=['culmen_length_mm','culmen_depth_mm'],inplace=True)
```

```python
len(df)
```

```python
len(df.loc[df['sex'].isnull()])
```

```python
df.loc[df['sex']=='.','sex'] = np.nan
```

```python
df.isnull().sum()
```

# filling null value with machine learning will lead to DATA LEAKING

```python
df = df.sort_values(by='sex').reset_index(drop=True)
```

```python
df_x = df.drop('sex',axis=1)
df_y = df['sex']
df_x = pd.get_dummies(df_x,drop_first=True)
ready_df = pd.concat([df_x,df_y],axis=1)
train_data = ready_df.loc[ready_df.sex.notnull()]
new_data = ready_df.loc[ready_df.sex.isnull()]
```

```python
X = train_data.drop('sex',axis=1)
y= train_data['sex']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
model = SVC()
scaler = StandardScaler()
########################################################
param_grid = {'svm__C':[0.1,1,3,5,10,20,30,100,250,1000,5000],'svm__kernel':['rbf','poly'],'svm__degree':[2,3,4,5]}
```

```python
# import pipeline
from sklearn.pipeline import Pipeline
```

```python
# create operation
ops = [('scale',scaler),('svm',model)]
pipe= Pipeline(ops)
```

```python
search_svm = GridSearchCV(pipe,param_grid=param_grid)
search_svm.fit(X_train,y_train)
```

```python
yhat_svm = search_svm.predict(X_test)
print(classification_report(y_test,yhat_svm))
```

```python
X_newdata = new_data.drop('sex',axis=1)
prediction = search_svm.predict(X_newdata)
prediction
```

```python
new_data.sex = prediction
```

```python
#concat to original data
encoded_df = pd.concat([train_data,new_data],axis=0)
```

```python
#function to report and plot trained decision tree
def dtree_report(tree,y_test,y_hat,X):
    print(classification_report(y_test,y_hat))
    
    plt.figure(figsize=(12,8),dpi=200)
    plot_tree(decision_tree=tree,feature_names=X_train.columns,filled=True);
```

```python
dtree = DecisionTreeClassifier(criterion='gini',min_samples_split=.3, max_features=3)
dtree.fit(X_train,y_train)
yhat_dtree = dtree.predict(X_test)
dtree_report(dtree,y_test,yhat_dtree,X)
```
