```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
pop_data = pd.read_csv('pop_data_001.csv')
user_data = pd.read_csv('user_data_001.csv')
new_df = pd.merge(user_data, pop_data)
```

```python
#classifiying rural and urban location based on population proper
new_df['location'] = new_df.population_proper.apply(lambda x: 'rural' if x <100_000 else 'urban')
display(new_df.head())
```

```python
age = new_df['age']
sns.distplot(age)
plt.show()
```

```python
location_and_age = new_df.groupby('location').age.mean()
print(location_and_age)
```

```python
sns.barplot(data=new_df, x= 'location', y='age')
plt.show()

sns.violinplot(x="location", y="age", data=new_df)
 
plt.show()

sns.set_style("darkgrid")
sns.set_palette("bright")
sns.despine()
 
sns.regplot(x="population_proper", y="age", data=new_df)
 
# Paste code to change the axes:
ax = plt.subplot()
ax.set_xticks([100000, 1000000, 2000000, 4000000, 8000000])
ax.set_xticklabels(["100k", "1m", "2m","4m", "8m"])

# Paste code to title the axes and the plot: 
ax.set_xlabel("City Population")
ax.set_ylabel("User Age")
plt.title("Age vs Population")
 
plt.show()
```
