```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.arange(0,101,)
b = a**2

fig = plt.figure()
#large
ax1 = fig.add_axes([0,0,1,1])
ax1.plot(a,b)

ax2 = fig.add_axes([.1,.65,.3,.3])
ax2.plot(b,a)
plt.show()
```

```python
fig = plt.figure(figsize=(10,3),dpi=200)
ax1 = fig.add_axes([0,0,1,1])

ax1.scatter(a,b)
```
