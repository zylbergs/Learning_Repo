```python
from matplotlib import pyplot as plt
import numpy as np
```

### matplotlib visualization OO methods

```python
x = np.random.randint(0,121,100)
y = x**5
```

```python
fig = plt.figure() #create object canvas
ax1 = fig.add_axes([0,0,1,1]) #create axes
ax2 = fig.add_axes([1,1,.5,.5]) 
#plotting section
ax1.plot(x,y)
#object attribute/axes attribute
ax1.set_xlabel('random 100 number')
ax1.set_ylabel('x pow 5')
ax1.set_title('full graph')

ax2.plot(x,y)
ax2.set_xlim(100,120)
#ax2.set_ylim(-2,2)
ax2.set_xlabel('x zoomed')
ax2.set_ylabel('y zoomed')
ax2.set_title('zoomed graph')
plt.show()
```

### Another Version of OO methods

```python
x = np.linspace(0,100,101)
y = pow(x,3)
```

```python
fig,axes = plt.subplots(2,2)
#all of figure argument can be stated to plt subplot 


axes[0][0].plot(x,y, label='old version')
axes[0][0].plot(x,y*2,label='new version')
axes[0][0].set_xlabel('x')
axes[0][0].set_ylabel('y= x pow 3')
axes[0][0].set_title('x and y pow 3')
#axes[0][0].legend(loc=(1.1,0.5))

axes[0][1].plot(x**2,y)
axes[0][1].set_xlabel('x pow 2')
axes[0][1].set_ylabel('y= x pow 3')
axes[0][1].set_title('x pow and y pow 3')


axes[1][0].plot(x*2,y)
axes[1][0].set_xlabel('x times 2')
axes[1][0].set_ylabel('y= x pow 3')
axes[1][0].set_title('x times 2 and y pow 3')

axes[1][1].plot(x*5,y**2)
axes[1][1].set_xlabel('x times 5')
axes[1][1].set_ylabel('y= x pow 6')
axes[1][1].set_title('x times 5 y pow 6')


fig.suptitle('random thoughts')

fig.tight_layout()

plt.show()
```

```python
trex_bone_mean = 18.7
trex_std = 1.8

bracio_bone_mean = 27.3
bracio_std = 3.2

trex_data = np.random.normal(trex_bone_mean,trex_std, size=10000)
bracio_data = np.random.normal(bracio_bone_mean,bracio_std, size =10000)

plt.hist(trex_data,bins=50)
plt.hist(bracio_data, bins=50) #create histogram or 
plt.xlabel('the bone length') #giving horizontal label
plt.ylabel('dino values') #giving vertical label
plt.title('trex vs bracio bones') #naming the graph datadeviation concept = to measure how spread is out data, and how normal is a data according to +- standard deviation

denomial concept = predict the success rate percentage in certain experiment
case = we want to conduct a donation to our cause. a person sent 500 email with an expectation 25 of them read and make a donation (5%). this experiment cycle again and again until 10000 times
we want to create a success rate prediction of person would donate.
plt.show() #this is for render the plot 
```

##### bars plot

##### stacked bars

```python
plt.close('all')

drinks = ["cappuccino", "latte", "chai", "americano", "mocha", "espresso"]
sales1 =  [91, 76, 56, 66, 52, 27]
sales2 = [65, 82, 36, 68, 38, 40]

plt.bar(range(len(drinks)),sales1)
plt.bar(range(len(drinks)), sales2, bottom=sales1)
plt.show()
```

##### side by side bars

```python
n =1 # dataset number 1
n2 = 2 # dataset number 2
t = 2 # amount of dataset
d = 6 # amout of bar pairs
w = 0.8 # plot widht

x = np.arange(len(drinks))

ax = plt.subplot()
x_values = [t* element + w*n for element in range(d)] #syntaxes for first dataset
x_values2 = [t*element +w*n2 for element in range(d)]
plt.bar(x_values, sales1, label='sales1')
plt.bar(x_values2, sales2, label= 'sales2')
ax.set_xticks([t*element + w*n for element in range(d)]) # this is the algo to match tick labels with bars
ax.set_xticklabels(drinks)
plt.legend()
print(x_values)
```

##### polar plots

```python
#data
skills = ['PYTHON','SQL','Data Visualization','Data Analyzation','Algorithm','English Language']
levels = [65,80,70,75,60,70,65] # to avoid misprecission model, add another index in the last list same as the value of first index

#creating figure n subplots
plt.figure(figsize=(10,8))
ax = plt.subplot(polar=True)

#creating base stats/theta
theta= np.linspace(0, 2*np.pi, len(levels))

#creating syntax for thetagrid model
lines,labels = plt.thetagrids(range(0,360, int(360/len(skills))),(skills))
#plotting levels
plt.plot(theta, levels)
plt.fill(theta, levels, color='#f2b418',alpha=0.5)

#plt.legend(['skills'],loc=0)
ax.set_facecolor('grey')
plt.savefig('my skill stats.png')
plt.show()
```

```python
months = range(12)
month_names = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
revenue = [16000, 14000, 17500, 19500, 21500, 21500, 22000, 23000, 20000, 19500, 18000, 16500]
y_lower = [(i -(i*0.1)) for i in revenue]
y_upper = [(i + (i*0.1)) for i in revenue]

#your work here
ax= plt.subplot()
plt.plot(months, revenue)
plt.fill_between(months, y_lower,y_upper, alpha=0.3)
ax.set_xticks(months)
ax.set_xticklabels(month_names)

plt.show()
```
