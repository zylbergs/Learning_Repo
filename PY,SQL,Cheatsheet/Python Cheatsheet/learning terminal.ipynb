```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
#data
skills = ['PYTHON','SQL','Data Visualizing','Analysis','Algorithm','Statistic','SEO']
levels = [70,80,75,75,60,65,20,70] # to avoid misprecission model, add another index in the last list same as the value of first index

#creating figure n subplots
plt.figure(figsize=(10,8))

ax = plt.subplot(polar=True)
plt.ylim(0,100)
#creating base stats/theta
theta= np.linspace(0, 2*np.pi, len(levels))

#creating syntax for thetagrid model
lines,labels = plt.thetagrids(range(0,360, int(360/len(skills))),(skills))
#plotting levels
plt.plot(theta, levels)
plt.fill(theta, levels, color='#f2b418',alpha=0.5)

#plt.legend(['skills'],loc=0)
ax.set_facecolor('grey')
plt.xticks(fontsize=16)
plt.savefig('my skill stats.png')
plt.show()
```
