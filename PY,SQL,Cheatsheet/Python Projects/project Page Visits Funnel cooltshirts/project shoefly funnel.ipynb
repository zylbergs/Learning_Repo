```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
visits = pd.read_csv('visits.csv', parse_dates=[1])
cart = pd.read_csv('cart.csv',parse_dates=[1])
checkout = pd.read_csv('checkout.csv',parse_dates=[1])
purchase = pd.read_csv('purchase.csv',parse_dates=[1])
```

```python
display(visits.head())
display(cart.head())
display(checkout.head())
display(purchase.head())
```

```python
# correlation of visits table and cart table,
# how many users that visits shoefly but not proceeding to put shirts into cart

visits_and_cart = pd.merge(visits,cart,how='left').reset_index(drop=True)
display(visits_and_cart.head())
len_vc = len(visits_and_cart)

not_carting = visits_and_cart[visits_and_cart['cart_time'].isnull()].reset_index(drop=True)
display(not_carting.head())
len_nc = len(not_carting)
visit_nocart_percentage = round((float(len_nc / len_vc * 100)),2)
print(visit_nocart_percentage)
```

```python
cart_and_check = pd.merge(cart,checkout,how='left').reset_index(drop=True)
len_cc = len(cart_and_check)
display(cart_and_check.head())
print(len_cc)

cart_nocheck = cart_and_check[cart_and_check['checkout_time'].isnull()].reset_index(drop=True)
len_cn = len(cart_nocheck)
display(cart_nocheck.head())
print(len_cn)

cart_nocheck_percentage = round(float(len_cn/len_cc *100),2)
print(cart_nocheck_percentage)
```

```python
check_and_purchase = pd.merge(checkout,purchase,how='left').reset_index(drop=True)
#display(check_and_purchase.info())
len_checkandpurchase = len(check_and_purchase)

nullpurchase = check_and_purchase[check_and_purchase.purchase_time.isnull()]
len_nullpurhcase = len(nullpurchase)
#display(nullpurchase.info())

percent_checknopurchase = round(float(len_nullpurhcase / len_checkandpurchase *100),2)
print(percent_checknopurchase)
```

```python
#merging  all data
all_data= visits.merge(cart,how='left').merge(checkout,how='left').merge(purchase,how='left')
#all_data.info()
#calculating time from visits to purchase
all_data['time_to_purchase'] = \
    all_data.purchase_time - \
    all_data.visit_time
display(all_data.head())
```

```python
#all customer data who did all the funnel process
checkout_then_purchase = all_data[~all_data['checkout_time'].isnull() & ~all_data['purchase_time'].isnull()].reset_index(drop=True)
vis_to_pur = all_data.loc[(~all_data.checkout_time.isnull()) & (~all_data.purchase_time.isnull())].reset_index(drop=True)
#its both displaying same result
display(checkout_then_purchase.head())

average_time_users_making_purchase = vis_to_pur.time_to_purchase.mean()
print(average_time_users_making_purchase)
```

```python
funnel_step_insight = [visit_nocart_percentage,cart_nocheck_percentage,percent_checknopurchase]
print(funnel_step_insight)
```
