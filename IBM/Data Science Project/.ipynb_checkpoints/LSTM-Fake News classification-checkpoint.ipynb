```python
X = df.drop('label',axis=1)
y = df['label']

#inspect shape
print(f'Features shape: {X.shape}')
print(f'label shape: {y.shape}')
```

```python
#importing LSTM deep learning tools
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Embedding,Dense,LSTM,SimpleRNN
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow import keras
import numpy as np
```

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
nltk.download('stopwords')
```

```python
p_stem = PorterStemmer()

article_corpus = []
for i in range(0,len(X)):
    print(i)
    content = re.sub('[^a-zA-Z]',' ', X['text'][i])
    content = content.lower()
    content = content.split()
    
    content = [p_stem.stem(word) for word in content if not word in stopwords.words('english')]
    content = ' '.join(content)
    article_corpus.append(content)
    
```

```python
len(article_corpus)
```

```python
vocabularies = 10000
#onehot mapping from vocabularies index
oh_repr = [one_hot(words,vocabularies) for words in article_corpus]
oh_repr
```

```python
#word embedding
c = []
for i in range(len(X)):
    m=len(df['text'][i].split(','))
    c.append(m)

print('Maximum length of a sentence is : ',max(c))
print('Average length of a sentence is : ',sum(c)/len(c))
```

```python
#padding the one hot data to make it in the same length
sentence = 48
emb_data = pad_sequences(oh_repr,padding='post',maxlen=sentence)
```

```python
#set embedded feature representation
embedded_feature = 45
lstm_model = Sequential()
lstm_model.add(Embedding(vocabularies,embedded_feature,input_length=sentence))
lstm_model.add(LSTM(100,activation='relu'))
lstm_model.add(Dense(20,activation='relu'))
lstm_model.add(Dense(20,activation='relu'))
lstm_model.add(Dense(1,activation='sigmoid'))
```

```python
optimizer = keras.optimizers.Adam(learning_rate=0.03)
lstm_model.compile(optimizer=optimizer,loss='binary_crossentropy',metrics=['accuracy'])
lstm_model.summary()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_final = np.array(emb_data)
y_final = np.array(y)

X_train, X_val, y_train, y_val = train_test_split(X_final, y_final, test_size=200,train_size=2000, random_state=42)

print('class 1 total sample: {}'.format(sum(y_train == 1)))
print('class 0 total sample: {}'.format(sum(y_train == 0)))
```

```python
lstm_model.fit(X_train,y_train,
               validation_data=(X_val,y_val),
              epochs=5,
              batch_size=100)
```

```python
print('training complete')
```
